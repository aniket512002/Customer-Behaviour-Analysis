```python
import pandas as pd

# Load dataset (correct path)
df = pd.read_csv('../dataset/amazon_sales_data.csv')

# Preview
df.head()
```

```python
df.info()
df.columns
```

```python
df.columns = df.columns.str.strip().str.lower()
df.columns
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')
```

```python
df['rating'] = df['rating'].fillna(df['rating'].median())
df['rating_count'] = df['rating_count'].fillna(0)
```

```python
df['discount_category'] = pd.cut(
    df['discount_percentage'],
    bins=[0, 20, 40, 60, 80, 100],
    labels=['Low', 'Medium', 'High', 'Very High', 'Extreme']
)
```

```python
top_products = df.groupby('product_name')['rating'].mean().sort_values(ascending=False).head(10)
top_products
```

```python
reviews_by_category = df.groupby('category')['rating_count'].sum().sort_values(ascending=False)
reviews_by_category
```

```python
discount_analysis = df.groupby('discount_category')['rating_count'].sum()
discount_analysis
```

```python
avg_price = df.groupby('category')['actual_price'].mean().sort_values(ascending=False)
avg_price
```

```python
df.to_csv('final_clean_data.csv', index=False)
print("✅ File saved successfully")
```
